```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# 데이터 정의
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [150, 200, 180, 220, 250, 230]
```

```python
# 막대 그래프 생성
plt.figure(figsize=(10, 6))
plt.bar(months, sales, color='skyblue', label='Sales')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()    # 범례 표시
plt.grid(True)
plt.show()
```

```python
# 데이터 정의
labels = ['Product A', 'Product B', 'Product C', 'Product D']
sizes = [400, 300, 500, 100]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)
```

```python
# 간결한 디자인의 파이차트
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution - Simple Design')
plt.axis('equal')    # 파이차트 원형 유지
plt.show()
```

```python
# 복잡한 디자인의 파이차트
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Sales Distribution - Complex Design')
plt.axis('equal')
plt.show()
```

```python
# 데이터 정의
data = {'Year' : [2018, 2019, 2020, 2021], 'Visitors' : [12000, 18000, 23000, 30000]}
df = pd.DataFrame(data)
```

```python
# 선 그래프 생성
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Visitors', data=df, marker='o', label='Visitors')
plt.title('Annual Website Visitors')
plt.xlabel('Year')
plt.ylabel('Number of Visitors')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 데이터 정의
data = {'Year': [2018, 2019, 2020, 2021],
        'Product A': [1500, 2000, 2400, 2800],
        'Product B': [2300, 2700, 3000, 3500],
        'Product C': [1800, 2200, 2600, 3000]}
df = pd.DataFrame(data)
```

```python
# 복합차트 생성
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Product A', data=df, marker='o', label='Product A')
sns.lineplot(x='Year', y='Product B', data=df, marker='o', label='Product B')
sns.lineplot(x='Year', y='Product C', data=df, marker='o', label='Product C')
plt.title('Anual Product Sales')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 데이터 생성
np.random.seed(10)
data = np.random.randn(1000)
```

```python
# 히스토그램 생성
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 데이터 생성
np.random.seed(10)
data={
    'x' : np.random.randn(100),
    'y' : np.random.randn(100)
}
df = pd.DataFrame(data)
```

```python
# 산점도 생성
plt.figure(figsize=(10, 6))
sns.scatterplot(x='x', y='y', data=df)
plt.title('Scatter Plot of Random Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
# 데이터 생성
np.random.seed(10)
data = {
    'Category' : np.random.choice(['A', 'B', 'C', 'D'], 100),
    'Value' : np.random.randn(100)
}
df = pd.DataFrame(data)
```

```python
# 박스플롯 생성
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=df)
plt.title('Box Plot of Random Data by Category')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```
